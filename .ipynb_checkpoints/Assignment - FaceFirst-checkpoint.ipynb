For performing the training taks, the images for the 4 categories were arranged in such a way where there were only 4 folders inside which all the images of that respective  class was placed.  This was done using some basic python where we move the images from the subfolders inside the 4 class folders to the parent folders.

As the images in the classes are not balanced, we are going to augment the images for some of the classes which has less images

#### Augmentation

```python
# from tensorflow import keras 

from tensorflow.keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img
datagen = ImageDataGenerator(shear_range=0.2, zoom_range=0.2,horizontal_flip=True) 

data_pth = "D:\FaceFirst\RealWorldOccludedFaces\images/"

for f in os.listdir(data_pth):
    q = input(f"\nAugment this class-{f}? ")
    if q in ["y","Y","yes"]:
        num = int(input(f"How many times each image should be augmented ? "))
        for image in os.listdir(data_pth+f):
            img = load_img(data_pth+f+"/"+image)  
            x = img_to_array(img) 
            # Reshape the input image 
            x = x.reshape((1, ) + x.shape)  
            i = 0
            for batch in datagen.flow(x, batch_size = 1, 
                      save_to_dir = data_pth+f,  
                      save_prefix ='cat', save_format ='jpg'):
                i += 1
                if i >= num: 
                    break
    else:
        continue
```

#### Loading Data and splitting

```python
from torch.utils.data.sampler import SubsetRandomSampler


data_pth = "D:\FaceFirst\RealWorldOccludedFaces\images"

def load_split_train_test(data_pth, valid_size=0.2):
    train_transforms = transforms.Compose([transforms.Resize((224,224)),
                                           transforms.ToTensor()])
    test_transforms = transforms.Compose([transforms.Resize((224,224)),
                                           transforms.ToTensor()])
    train_data = datasets.ImageFolder(data_pth,transform=train_transforms)
    test_data = datasets.ImageFolder(data_pth,transform=test_transforms)
    
    

    num_train = len(train_data)
    indices = list(range(num_train))
    split = int(np.floor(valid_size * num_train))
    np.random.shuffle(indices)
    
    train_idx, test_idx = indices[split:], indices[:split]
    train_sampler = SubsetRandomSampler(train_idx) 
    test_sampler = SubsetRandomSampler(test_idx)
 
    trainloader = torch.utils.data.DataLoader(train_data, sampler=train_sampler, batch_size=64)
    testloader = torch.utils.data.DataLoader(test_data, sampler=test_sampler, batch_size=64)
    return trainloader, testloader,test_data

trainloader, testloader,test_data = load_split_train_test(data_pth, .2)
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = models.resnet50(pretrained=True)
model
```

#### Model Parameters 

```python
for param in model.parameters():
    param.requires_grad = False
    
model.fc = nn.Sequential(nn.Linear(2048,512),
                        nn.ReLU(),
                        nn.Dropout(0.2),
                        nn.Linear(512,4),
                        nn.LogSoftmax(dim=1))
criterion = nn.NLLLoss()
optimizer = optim.Adam(model.fc.parameters(),  lr =0.003)
# Decay LR by a factor of 0.1 every 4 epochs
scheduler = lr_scheduler.StepLR(optimizer, step_size=4, gamma=0.1)
model.to(device)
```

### Training

```python
epochs=3
steps=0
running_loss = 0
print_every = 1
train_losses, test_losses = [],[]

best_model_wts = copy.deepcopy(model.state_dict())
best_acc = 0.0

for epoch in tqdm(range(epochs)):
    for inputs, labels in (trainloader):
        steps += 1
        inputs, labels = inputs.to(device), labels.to(device)
        optimizer.zero_grad()
        
        logps = model.forward(inputs)
        loss = criterion(logps,labels)
        loss.backward()
        optimizer.step()
        running_loss += loss.item()
        
        if steps % print_every==0:
            test_loss = 0
            accuracy = 0
            model.eval()
            with torch.no_grad():
                for test_inputs, test_labels in testloader:
                    test_inputs,test_labels = test_inputs.to(device), test_labels.to(device)
                    test_logps = model.forward(test_inputs)
                    batch_loss = criterion(test_logps, test_labels)
                    test_loss += batch_loss.item()
                    
                    ps = torch.exp(test_logps)
                    top_p, top_class = ps.topk(1,dim=1)
                    equals = top_class == test_labels.view(*top_class.shape)
                    accuracy += torch.mean(equals.type(torch.FloatTensor)).item()
                    
            train_losses.append(running_loss/len(trainloader))
            test_losses.append(test_loss/len(testloader))
            print(f"Epoch {epoch+1} ....."
                  f"Train Loss: {running_loss/print_every:.3f}.."
                  f"Test loss: {test_loss/len(testloader):.3f}.. "
                  f"Test accuracy: {accuracy/len(testloader):.3f}")
            
            if accuracy > best_acc:
                best_acc = accuracy
                best_model_wts = copy.deepcopy(model.state_dict())
                
            running_loss = 0
            model.train()
                    
        
```

Best Test Accuracy obtained is 92.0 %

Further training of the model for more epochs has caused overfitting of the model due to less number of images in some folders. Hence settled for the model at the 3rd epoch

### Inference

#### On Image ( Unseen images)

```python
import cv2
import time

test_img_dir = "D:/FaceFirst/RealWorldOccludedFaces/test_images/"


for classes_dir in os.listdir(test_img_dir):
    for image in os.listdir(test_img_dir+classes_dir):
        image = cv2.imread(test_img_dir+classes_dir+"/"+image)
        image_transform = transforms.Compose([transforms.ToPILImage(),
                                              transforms.Resize((224,224)),
                                                   transforms.ToTensor()])

        image_tensor = image_transform(image)
        
        input_batch = image_tensor.unsqueeze(0)

        # move the input tensor and model to the computation device
        input_batch = input_batch.to(device)
        model.to(device)

        with torch.no_grad():
            start_time = time.time()
            output = model(input_batch)
            end_time = time.time()

        # get softmax prob
        prob = torch.nn.functional.softmax(output[0],dim=0)

        top_prob, top_prob_cat = torch.topk(prob,1)

        pred_label = [i for i in class2idx if class2idx[i]==top_prob_cat[0]]

#         image = cv2.putText(image, f"{top_prob[0].item()*100:.3f}%", (15,30), cv2.FONT_HERSHEY_SIMPLEX,
#                        0.6, (0, 255, 0), 2, cv2.LINE_AA)
#         image = cv2.putText(image, f"{pred_label[0]}", (15,60), cv2.FONT_HERSHEY_SIMPLEX,
#                         0.6, (0, 255, 0), 2, cv2.LINE_AA)

        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        imgplot = plt.imshow(image)
        plt.show()

        print(f"Predicted as {pred_label[0]} with {top_prob[0]} confidence") 

        print(f"Forward pass time: {(end_time-start_time):.3f} seconds")
```

#### On Video

```python
vid_dir = "D:/FaceFirst/RealWorldOccludedFaces/test_video/FaceDetection2.mp4"

cap = cv2.VideoCapture(vid_dir)
cap.set(cv2.CAP_PROP_BUFFERSIZE,1)
# get the frame width and height
frame_width = int(cap.get(3))
frame_height = int(cap.get(4))

out = cv2.VideoWriter(f"RealWorldOccludedFaces/test_video/result.mp4", 
                      cv2.VideoWriter_fourcc(*'MP4V'), 20.0, 
                      (frame_width, frame_height))

# to count the total number of frames iterated through
frame_count = 0
# to keep adding the frames' FPS
total_fps = 0
```

```python
while(cap.isOpened()):
    ret, frame = cap.read()
    if ret:
        rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        image_transform = transforms.Compose([transforms.ToPILImage(),
                                          transforms.Resize((224,224)),
                                               transforms.ToTensor()])
        input_tensor = image_transform(rgb_frame)
        input_batch = input_tensor.unsqueeze(0)
        
        input_batch = input_batch.to(device)
        model.to(device)
        
        with torch.no_grad():
            start_time = time.time()
            output = model(input_batch)
            end_time = time.time()
        
        prob = torch.nn.functional.softmax(output[0], dim=0)
        # get the top 1 prediction
        top_prob, top_cat = torch.topk(prob, 1)
        pred_label = [i for i in class2idx if class2idx[i]==top_cat[0]]
        
        # get current fps
        fps = 1/ (end_time - start_time)
        total_fps += fps
        frame_count += 1
        img = cv2.putText(frame, f"{fps:.3f} FPS", (15, 30), cv2.FONT_HERSHEY_SIMPLEX,
                        2, (0, 255, 0), 2)
        img = cv2.putText(frame, f"{pred_label[0]}", (15, 90), cv2.FONT_HERSHEY_SIMPLEX,
                        2, (0, 255, 0), 2)
        cv2.imshow('Result', frame)
        out.write(frame)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    else:
        break
cap.release()
out.release()

cv2.destroyAllWindows()

avg_fps = total_fps / frame_count
print(f"Average FPS: {avg_fps:.3f}")
            
```

```python
ort_provider = ['CPUExecutionProvider']
if torch.cuda.is_available():
    model.to('cuda')
    ort_provider = ['CUDAExecutionProvider']

ort_sess = ort.InferenceSession(model_fp32_path, providers=ort_provider)
```

#### Evaluation on test Set

```python
correct_pt = 0
correct_onnx = 0
tot_abs_error = 0
total = 0

for img_batch, label_batch in tqdm(testloader, ascii=True, unit="batches"):
    ort_inputs =  {ort_sess.get_inputs()[0].name: to_numpy(img_batch)}
    ort_outputs = ort_sess.run(None, ort_inputs)[0]
    
    ort_preds = np.argmax(ort_outputs, axis=1)
    correct_onnx += np.sum(np.equal(ort_preds, to_numpy(label_batch)))
    
    if torch.cuda.is_available():
        img_batch = img_batch.to("cuda")
        label_batch = label_batch.to("cuda")
     
    with torch.no_grad():
        pt_outs = model(img_batch)
        
    total += label_batch.size(0)
    pt_preds = torch.argmax(pt_outs, dim=1)
    correct_pt += torch.sum(pt_preds == label_batch).item()
    
    tot_abs_error += np.sum(np.abs(to_numpy(pt_outs) - ort_outputs))
    
print(f"pt top-1 acc = {100.0 * correct_pt/(total)} with {correct_pt} correct samples")
print(f"onnx top-1 acc = {100.0 * correct_onnx/(total)} with {correct_onnx} correct samples")

mae = tot_abs_error/(1000*(total))
print(f"mean abs error = {mae} with total abs error {tot_abs_error}")
    
        
```

We can see an accuracy of 92.5 % for the 730 samples of the test set

#### Quantization

```python
torch.cuda.empty_cache()
```

```python
model_prep_path = 'resnet50_prep.onnx'

quantization.shape_inference.quant_pre_process(model_fp32_path, model_prep_path, skip_symbolic_shape=False)
```

```python
class QuntizationDataReader(quantization.CalibrationDataReader):
    def __init__(self, torch_ds, batch_size, input_name):

        self.torch_dl = torch.utils.data.DataLoader(torch_ds, batch_size=batch_size, shuffle=False)

        self.input_name = input_name
        self.datasize = len(self.torch_dl)

        self.enum_data = iter(self.torch_dl)

    def to_numpy(self, pt_tensor):
        return pt_tensor.detach().cpu().numpy() if pt_tensor.requires_grad else pt_tensor.cpu().numpy()

    def get_next(self):
        batch = next(self.enum_data, None)
        if batch is not None:
            return {self.input_name: self.to_numpy(batch[0])}
        else:
            return None

    def rewind(self):
        self.enum_data = iter(self.torch_dl)

qdr = QuntizationDataReader(test_data, batch_size=64, input_name=ort_sess.get_inputs()[0].name)
```

```python
q_static_opts = {"ActivationSymmetric":False,
                 "WeightSymmetric":True}
if torch.cuda.is_available():
    q_static_opts = {"ActivationSymmetric":True,
                  "WeightSymmetric":True}

model_int8_path = 'resnet50_int8.onnx'
quantized_model = quantization.quantize_static(model_input=model_prep_path,
                                               model_output=model_int8_path,
                                               calibration_data_reader=qdr,
                                               extra_options=q_static_opts)
```

```python
model_int8_path = 'resnet50_int8.onnx'
```

#### Evlauation on test set after ONNX conversion and quantization from FP32 to INT8

```python
ort_int8_sess = ort.InferenceSession(model_int8_path, providers=ort_provider)
```

```python
correct_int8 = 0
correct_onnx = 0
tot_abs_error = 0
total = 0

for img_batch, label_batch in tqdm(testloader, ascii=True, unit="batches"):
    ort_inputs = {ort_sess.get_inputs()[0].name: to_numpy(img_batch)}
    ort_outs = ort_sess.run(None, ort_inputs)[0]

    ort_preds = np.argmax(ort_outs, axis=1)
    correct_onnx += np.sum(np.equal(ort_preds, to_numpy(label_batch)))
    
    ort_int8_outs = ort_int8_sess.run(None, ort_inputs)[0]
    ort_int8_preds = np.argmax(ort_int8_outs, axis=1)
    correct_int8 += np.sum(np.equal(ort_int8_preds, to_numpy(label_batch)))
    
    tot_abs_error += np.sum(np.abs(ort_int8_outs - ort_outs))
    total += label_batch.size(0)
    
print(f"onnx top-1 acc = {100.0 * correct_onnx/total} with {correct_onnx} correct samples")
print(f"onnx int8 top-1 acc = {100.0 * correct_int8/total} with {correct_int8} correct samples")

mae = tot_abs_error/(1000*total)
print(f"mean abs error = {mae} with total abs error {tot_abs_error}")
```

Here the Accuracy for the int8 model has reduced about 4 % from the accuracy of the fp32 onnx model after quantization
